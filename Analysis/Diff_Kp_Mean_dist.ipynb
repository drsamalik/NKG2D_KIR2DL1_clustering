```python
import os
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.optimize import least_squares
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['axes.linewidth'] = 4
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=19, length=8, width=4)
lss = [(0,(1,0)),(0, (1, 0)), (0,(1,0)), (0, (3, 1, 1, 1))]
m_size = 15
lw = 4.0
def Plotting(ARL,mean_pVav,std_pVav,cond,marker,ls,color):
    plt.errorbar(
        ARL, mean_pVav,
        fmt=marker, markersize=m_size, markerfacecolor='lightgray',
        markeredgewidth=3.0, markeredgecolor=color,
        elinewidth=3.5, capsize=10, capthick=20,
        ecolor=color, alpha=0.99, 
        color=color, label=cond )#, lw=lw, ls=ls, label=cond  # Combine line and marker in legend
    #)
current_path = os.getcwd()  
folder_name = ['Ulbp_40']
at_time = np.arange(0,41,5)
Kp_condns = ['Kp_Both','No_Kp_Inh','No_Kp_Act','No_Kp']
clus_conds = ['Disjoint','Homogeneous','Overlap']
HLA_condns = ['Low_HLA','High_HLA']
kp_conds = ['KP','no KP inh','no KP act','no KP']
markers = ['s','o','^','<']
colors = ['blue','gray','red','black']
for k,cond in enumerate(Kp_condns):
    file_path = f'{current_path}/{cond}/{condition[0]}/{HLA_condns[0]}/{folder_name[0]}/pVav_number_w_time.csv'
    df_pVav_ensemble_time = pd.read_csv(file_path, index_col=0)
    mean_pVav = []
    std_pVav = []
    for j,i in enumerate(at_time):
        pVav = (df_pVav_ensemble_time[f'time_{i}'].values)
        mean_pVav.append(pVav.mean())
        std_pVav.append(pVav.std())
    Plotting(at_time,mean_pVav,std_pVav,kp_conds[k],markers[k],lss[0],colors[k])
t_size =30
plt.xticks([0,10,20,30,40],fontname="Arial",fontsize = t_size)
plt.yticks([0,8,16],fontname="Arial",fontsize = t_size)
plt.ylim(-1,18.5)
plt.legend(bbox_to_anchor=(0.0, 1),fontsize=17, loc='upper left')
plt.savefig('Disjoint_Low_HLA_mean_pVav_vs_time.png')
plt.show()
```
