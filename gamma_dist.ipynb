```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for the Gamma distribution
shape_param = 33.97  # α (also called 'k' or 'a')
scale_param = 1.62  # β (scale is 1/β in scipy)

# Generate x values
x = np.linspace(0, 100, 1000)

# Compute the PDF
pdf_values = gamma.pdf(x, a=shape_param, scale=1/scale_param)

# Plot the Gamma distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, label=f'Gamma PDF (α={shape_param}, β={1/scale_param})', color='b')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Gamma Distribution')
plt.legend()
plt.grid()
plt.show()
```
