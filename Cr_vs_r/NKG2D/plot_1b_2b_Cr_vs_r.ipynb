```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_1b_125 = pd.read_csv('fig_1b_random_all.csv', index_col=0)
columns = df_1b_125.columns
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=14, length=6, width=3)
plt.rcParams['axes.linewidth'] = 3
plt.rcParams['font.family'] = 'Arial'
m_size = 8
plt.plot(0.04*df_1b_125[columns[0]], df_1b_125[columns[1]]/df_1b_125[columns[1]][0], linestyle='-',marker='o',markersize=m_size, color='lightpink',label = 'NKG2D Image',alpha=0.9)
# for i in columns[2:]:
#     plt.plot(0.04*df_1b_125[columns[0]], df_1b_125[i]/df_1b_125[i][0], linestyle='-', color='gray',alpha=0.14)
avg_coor_1b = df_1b_125.loc[:, 'iter0':'iter9'].mean(axis=1)
plt.plot(0.04*df_1b_125[columns[0]], avg_coor_1b/avg_coor_1b[0], linestyle='-',marker='d', color='red',label = 'Randomized Mean',alpha=0.5)
plt.axhline(y=0, color='gray', linestyle='--',lw=3.,alpha=0.35)
# plt.xlabel('r',fontsize=25)
# plt.ylabel('C(r)',fontsize=25)
plt.xticks([0.0,0.3,0.6],fontsize=20)
plt.yticks([0.0,0.5,1.0],fontsize=20)
plt.xlim(-0.02,0.7)
plt.legend(bbox_to_anchor=(0.39, 1.0),fontsize=16, loc='upper left')
plt.savefig('NKG_Cr_vs_r.png')
plt.show()
```
